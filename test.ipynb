```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/decoding-gaming-trends/train.csv')
data.head()
```

```python
data['age'] = 2025-data['Release Year']
data['Revenue_PP'] = data['Revenue (Millions $)']/data['Players (Millions)']
data.head()
```

```python
plt.scatter(data['age'],data['Players (Millions)'])
```
